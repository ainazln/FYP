```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv") #Reading the dataset in a dataframe using Pandas
df.head()
```

```python
# dropping passed columns 
df1.drop(["Title", "Shop1", "Author", "Shop2"], axis = 1, inplace = True) 
df1.head()
```

```python
#rename column
df1.rename(columns = {'Time':'time',  '_1yktim':'title', 'Rating':'rating'}, inplace = True)
df1.head()
```

```python
#save file
df1.to_csv("dataclean.csv")
```

```python
import pandas as pd
df2 = pd.read_csv('dataclean.csv') 
  
#Data preprocessing  
import nltk
nltk.download('wordnet')
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 

def preprocess(sentence):
    sentence=str(sentence)
    #lowercase text
    sentence = sentence.lower()
    #remove chinese character
    sentence = re.sub("([^\x00-\x7F])+"," ",sentence)
    #remove symbols and emoji
    cleanr = re.compile('<.*?>')
    #remove white space
    cleantext = re.sub(cleanr, '', sentence)
    #remove number
    rem_num = re.sub('[0-9]+', '', cleantext)
    #tokenie the sentence
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num) 
    #remove stop word
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    return " ".join(filtered_words)


df2['newReview']=df2['review'].map(lambda s:preprocess(s)) 
df2.head(30)
```

```python
#drop null value
df2.dropna(subset=['review'], inplace=True)
df2.head()
```

```python
# dropping passed columns 
df2.drop(["Unnamed: 0"], axis = 1, inplace = True) 
df2.head()
```

```python
df2.to_csv("newdataclean.csv")
```
